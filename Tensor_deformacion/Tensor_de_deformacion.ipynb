```python
archivo = genfromtxt("antofagasta.dat",delimiter="")

x = archivo[:,0]
y = archivo[:,1]
ux = archivo[:,2]
uy = archivo[:,4]
vals = zeros(31)
vects = zeros((31,4))
r = zeros((31,2))

#31
for n in range(31):
    m = zeros((2,2))
    m[0,0] = 0.5*( ux[n]/x[n] + ux[n]/x[n] )
    m[1,1] = 0.5*( uy[n]/y[n] + uy[n]/y[n] )
    m[0,1] = 0.5*( ux[n]/y[n] + uy[n]/x[n] )
    m[1,0] = m[0,1]
    
    eigenvalues, eigenvectors = eig(m)
    
    if eigenvalues[0] > 0:
        vals[n] = eigenvalues[0]
        vects[n,0] = x[n]
        vects[n,1] = y[n]
        vects[n,2] = x[n] + eigenvectors[:,0][0]*eigenvalues[0]
        vects[n,3] = y[n] + eigenvectors[:,0][1]*eigenvalues[0]
        r[n,0] = eigenvectors[:,0][0]
        r[n,1] = eigenvectors[:,0][1]
        
    else:
        vals[n] = eigenvalues[1]
        vects[n,0] = x[n]
        vects[n,1] = y[n]
        vects[n,2] = x[n] + eigenvectors[:,1][0]*eigenvalues[1]
        vects[n,3] = y[n] + eigenvectors[:,1][1]*eigenvalues[1]
        r[n,0] = eigenvectors[:,1][0]
        r[n,1] = eigenvectors[:,1][1]
        
print vals
```

```python
X,Y,U,V = zip(*vects)
figure(figsize=(13,10))
ax = gca()
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=100)
ax.set_xlim([-71.5,-67])
ax.set_ylim([-25.5,-22.5])
xlabel(u'Latitud',fontsize=15)
ylabel(u'Longitud',fontsize=15)
draw()
show()
```
