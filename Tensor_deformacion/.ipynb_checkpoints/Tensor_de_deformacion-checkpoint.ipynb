```python
archivo = genfromtxt("antofagasta.dat",delimiter="")

#Se convirtio todo a metros primero segun la formula x = theta(rads)*radio, usando rad promedio y ecuador
lat = archivo[:,0]
lon = archivo[:,1]
x = lat*pi*6371000.0/180.0
y = lon*pi*6378000.0/180.0
ux = archivo[:,2]
uy = archivo[:,4]
vals = zeros(31)
vects = zeros((31,4))
r = zeros((31,2))

#31
for n in range(31):
    m = zeros((2,2))
    m[0,0] = 0.5*( ux[n]/x[n] + ux[n]/x[n] )
    m[1,1] = 0.5*( uy[n]/y[n] + uy[n]/y[n] )
    m[0,1] = 0.5*( ux[n]/y[n] + uy[n]/x[n] )
    m[1,0] = m[0,1]
    
    eigenvalues, eigenvectors = eig(m)
    
    if eigenvalues[0] > eigenvalues[1]:
        vals[n] = eigenvalues[0]
        vects[n,0] = lat[n]
        vects[n,1] = lon[n]
        vects[n,2] = lat[n] + eigenvectors[:,0][0]*100
        vects[n,3] = lon[n] + eigenvectors[:,0][1]*100
        r[n,0] = eigenvectors[:,0][0]
        r[n,1] = eigenvectors[:,0][1]
        
    else:
        vals[n] = eigenvalues[1]
        vects[n,0] = lat[n]
        vects[n,1] = lon[n]
        vects[n,2] = lat[n] + eigenvectors[:,1][0]*100
        vects[n,3] = lon[n] + eigenvectors[:,1][1]*100
        r[n,0] = eigenvectors[:,1][0]
        r[n,1] = eigenvectors[:,1][1]
        
print vals
```

```python
X,Y,U,V = zip(*vects)
figure(figsize=(13,10))
ax = gca()
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=100)
ax.set_xlim([-73.5,-66])
ax.set_ylim([-26,-22.5])
xlabel(u'Latitud',fontsize=15)
ylabel(u'Longitud',fontsize=15)
draw()
show()
```

### Si se quiere en la siguiente celda se puede fabricar un archivo que contiene la posicion de la estacion (lat,long), el valor propio principal asociado y el vector propio(partiendo desde el origen con coordenadas x,y)

```python
file = open("valoresyvectores.dat", "w")

file.write("#lat" + "\t" + "long" + "\t" + "valor" + "\t" + "eje-x" + "\t" + "eje-y" + "\n")
for i in range(31):
    file.write( str(lat[i]) + "\t" + str(lon[i]) + "\t" + str(vals[i]) + "\t" + str(r[i,0]) + "\t" + str(r[i,1]) +"\n" )

file.close()
```

### Estos resultados pueden ser interpretados de la siguiente manera, el vector propio indica el sentido en el cual las posiciones sobre el mapa tuvieron un corrimiento en el lugar especificado, y el valor propio indica cuanto se corrió el lugar.

### A continuacion los vectores propios de cada Tensor, especificado por la latitud y longitud de la estacion en cuestion. El el eje x tambien se muestra el valor propio principal (el mayor entre los dos)

```python
figure(figsize=(25,80))

for i in range(31):
    subplot(11,3,i+1)
    plot()
    X,Y,U,V = [0,0,r[i,0],r[i,1]]
    ax = gca()
    ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=100)
    axis("equal")        
    title("Latitud: "+str(lat[i])+" , longitud: "+str(lon[i]))
    xlabel(u'x (eigval= '+str(vals[i])+')')
    ylabel(u'y')
subplots_adjust(hspace=.5)   
show()
```
